```python
import pandas as pd
```

```python
#import the csv and input it into the dataframe
sneaker_path = "../static/data/StockX-Data-Contest-2019-3.csv"
sneaker_df = pd.read_csv(sneaker_path)
sneaker_df.head()
```

```python
#check for any null value in the dataframe
sneaker_df.isnull().values.any()
```

```python
#check the dataframe size
sneaker_df.shape
```

```python
#check the dataframe column
sneaker_df.columns
```

```python
#filter only off white sneakers and input them into a new dataframe
off_white_df= sneaker_df[sneaker_df.Brand.str.contains('Off-White',case=False)]
off_white_df.head()
```

```python
#remove all commas and dollar signs in the Retail Price and Sale Price columns
off_white_df["Retail Price"] = off_white_df["Retail Price"].str.replace(',','')
off_white_df["Retail Price"] = off_white_df["Retail Price"].str.replace('$','')
off_white_df["Sale Price"] = off_white_df["Sale Price"].str.replace(',','')
off_white_df["Sale Price"] = off_white_df["Sale Price"].str.replace('$','')
off_white_df.head()
```

```python
off_white_df["Retail Price"] = off_white_df["Retail Price"].astype(int)
off_white_df["Sale Price"] = off_white_df["Sale Price"].astype(int)
```

```python
#create a new column and calculate the sneaker value
off_white_df["Gain/Loss"] = off_white_df["Sale Price"] - off_white_df["Retail Price"]
off_white_df.head()
```

```python
#check off white dataframe size
off_white_df.shape
```

```python
#see the basic data summary of the offwhite data frame
off_white_df.describe()
```

```python
#export as a csv file and save it into a data folder
off_white_df.to_csv(r'../static/data/off_white_2019.csv', index = False)
```
